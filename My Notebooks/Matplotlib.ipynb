```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,5,12)
y=x**2
```

```python
# Functional method
```

```python
plt.plot(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary Trend By Age')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'-r')
plt.subplot(1,2,2)
plt.plot(y,x,'-b')
```

```python
# OBJECT ORIENTED METHOD
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('Age')
axes.set_ylabel('Salary')
axes.set_title('Age vs Salary')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

# MATPLOTLIB 2


```python
fig,axes=plt.subplots()
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=2)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
#Figure Size and DPI
```

```python
fig=plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x square')
ax.plot(x,x**3,label='x cube')
ax.legend(loc=10)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```
