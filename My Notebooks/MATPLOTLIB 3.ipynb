```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1=np.linspace(5,25,5)
y1=x1**2
```

```python
fig1=plt.figure()
ax1=fig1.add_axes([2,2,1,1])
ax1.plot(x1,y1,color='green',linewidth=1,alpha=0.5,linestyle='-.',marker='o',
        markersize=20,markerfacecolor='yellow',markeredgewidth=3,
         markeredgecolor='red')
ax1.set_xlim([0,50])
ax1.set_ylim([0,1000])
```

```python
plt.scatter(x1,y1)
```
